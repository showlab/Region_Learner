```python
import json
import numpy as np
from matplotlib import pyplot as plt
import decord
# import torch.nn.functional as F
from torchvision import transforms
from PIL import Image
import random
```

```python
def load_vis_re(result_pth):
    with open(result_pth, 'r') as f:
        re = json.load(f)
    return re

def img_plot(img):
    img = img.resize((224,224))
#     plt.imshow(img)
    return img
    
    
def clu_plot(img):
#     plt.imshow(img)
    return img

decord.bridge.set_bridge("torch")
def read_frames_decord(video_path, frame_idx=None):
    video_reader = decord.VideoReader(video_path, num_threads=1)
    vlen = len(video_reader)
    frame_idx = vlen//2
    frame = video_reader[frame_idx]
    frame = frame.float() / 255
    frame = frame.permute(2, 0, 1)
#     print(type(frame), frame.shape)
    frame = transforms.ToPILImage()(frame).convert("RGB")
    return frame, frame_idx
```

```python
######## This is visualization script #########
# get tmp result from pre-trained models
WebVid_root="/cfs/cfs-4260a4096/260a4096/445-mds11/WebVid/val"
result_pth = 'tmp_result/vis/1107_010021/vis.json'
re = load_vis_re(result_pth)
keys = list(re.keys()) # key is video id

# get random result item
# print(keys)

def get_one_sample(rand_id):
    # rand_id = 7
    fr = 1
    vid = keys[rand_id]
    data_item = re[vid]

    # get each var from item (e.g., cluster id and region mask)
    clu_re = data_item['cluster_id']
    clu_re = np.array(clu_re)#.transpose(1,0)
    clu_re = np.flip(clu_re)
    raw_caption = data_item['raw_caption']
    frame_idxs = data_item['frame_idxs']
    region_mask = data_item['region_mask']


    img_list = []
    img_name_list = []

    # plot raw frame
    frame_name = "%s/frame_%s"%(vid, frame_idxs)
    frame_name = frame_name.replace('.mp4','')
    frame_name = frame_name.replace('/','_')
    
    frame, _ = read_frames_decord("%s/%s"%(WebVid_root,vid), frame_idxs)
    img_list.append(img_plot(frame))
    img_name_list.append(frame_name)

    # plot cluster
    img_list.append(clu_plot(clu_re))
    img_name_list.append(frame_name+"_clu")

    # plot region mask
    region_mask=np.array(region_mask)#.transpose(1,2)
#     print(region_mask.shape)
    # print(len(region_mask))
    # print(region_mask[0].shape)
    for i in range(len(region_mask)):
        region = np.flip(region_mask[i])#.transpose(1,0)
    #     region = np.flip(region,1)
        img_list.append(clu_plot(region))
        img_name_list.append(frame_name+"_region_%d"%(i))
    return img_list, img_name_list, raw_caption

def plot_save_img(img, img_name):
    plt.figure()
    plt.imshow(img)
    plt.savefig(img_name+".jpg")
    
def plot_save_clu(img, img_name):
    plt.figure()
    plt.imshow(img, plt.get_cmap('plasma'))
    plt.savefig(img_name+".jpg")
    
```

```python
rand_id = random.randint(1, len(keys))
img_list, img_name_list, caption = get_one_sample(rand_id)
print(caption)
plot_one_sample(img_list, img_name_list)
# plot_save_img(img_list, img_name_list)
```

```python
# print(len(img_list))
def plot_one_sample(img_list, img_name_list):
    plt.figure(figsize=(20, 6))
    for i in range(1,len(img_list)+1):
        plt.subplot(1,len(img_list),i)
        if i==2:
            cmap=plt.get_cmap('plasma')
            plt.imshow(img_list[i-1], cmap)#binary,viridis
        else:
            plt.imshow(img_list[i-1])#binary,viridis
        plt.xticks([])
        plt.yticks([])  
    plt.show()

def plot_save_img(img_list, img_name_list):
    plt.figure()
    plt.xticks([])
    plt.yticks([])
    for i in range(1,len(img_list)+1):
        if i==2:
            cmap=plt.get_cmap('plasma')
            plt.imshow(img_list[i-1], cmap)#binary,viridis
        else:
            plt.imshow(img_list[i-1])#binary,viridis
#         plt.show()
        plt.savefig("./vis_results/"+img_name_list[i-1]+".jpg")
    
```
